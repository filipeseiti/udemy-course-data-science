```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.set_style('ticks')
#plt.figure(figsize=(12,8))
sns.countplot(x='sex', data=tips)
#sns.despine(left=True)
```
